```python
import random
import urllib.request, json

print('Тесты по психологии для удобства прорешивания поделены на 2 последовательные части.')
start = int(input('Введи цифру 1 или 2, соответствующую номеру части для прорешивания: '))
print()
print('Отлично, приступай! (В случае множественного ответа вводи буквы ответа слитно по алфавитному порядку)')
print()

if start == 1:
  url = 'https://drive.google.com/file/d/17WqBVwIqxgrTL9zksXFE5EJrgSmmYWLN/view?usp=sharing'
  url = 'https://drive.google.com/uc?id=' + url.split('/')[-2]
if start == 2:
  url = 'https://drive.google.com/file/d/1i18BFnwGK-HSykOMUxPw3zpcvuCYqhhi/view?usp=sharing'
  url = 'https://drive.google.com/uc?id=' + url.split('/')[-2]
if start == 3:
  url = 'https://drive.google.com/file/d/14y5hnlrAVRf9oH76qCWT-9hdEH4Q6ktl/view?usp=sharing'
  url = 'https://drive.google.com/uc?id=' + url.split('/')[-2]
if start == 4:
  url = 'https://drive.google.com/file/d/1KeQPBv0k_7-QqAp9b_zV2aDuwN0GSaAV/view?usp=sharing'
  url = 'https://drive.google.com/uc?id=' + url.split('/')[-2]

with urllib.request.urlopen(url) as url:
    js = json.load(url)

nums = [i for i in range(len(js))]

while nums != []:
  rand_index_of_num = random.randint(0, len(nums)-1)
  qq = nums.pop(rand_index_of_num)
  print(js[qq]['quest'])
  print('a)', js[qq]['a'])
  print('b)', js[qq]['b'])
  print('c)', js[qq]['c'])
  print('d)', js[qq]['d'])
  if len(js[qq]) > 6:
    print('e)', js[qq]['e'])
  student_answer = 'z'
  while student_answer != js[qq]['answer']:
    student_answer = input('Введи букву(ы) ответа: ').strip().lower()
    if student_answer == js[qq]['answer']:
      print('Верно :)')
      print()
    else:
      print('Неверно :(')
      print()

print('Молодец, ты прорешал все тесты!')
```
