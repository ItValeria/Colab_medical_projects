```python
import pandas as pd
url='https://drive.google.com/file/d/1XG56eToAu3jp4io-WXWdSWR0QN0lMRE0/view?usp=sharing'
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_csv(url)
df = df.iloc[:, :-1]
df = df.drop([1,2], axis = 0)
df = df.dropna()
while True:
    print()
    try:
        er = 0
        print("Введите номер действия:")
        print("1) Получение информации о преимуществах препарата")
        print("2) Получение расширенной информации о преимуществах препарата, включая производителя, лекформу, показание")
        print("3) Выход")
        k = int(input())
        if k==1:
            print(("Введите название препарата для получения информации о преимуществах:"))
            i = input().strip().lower()
            print()
            res = df[df['Продукт'].str.contains(i)]
            if len(res) > 0:
                for index, row in res.iterrows():
                    print(row['Продукт'], "-", row['Преимущество'])
            else:
                er += 1
            i = i[0].upper() + i[1:]
            res = df[df['Продукт'].str.contains(i)]
            if len(res) > 0:
                for index, row in res.iterrows():
                    print(row['Продукт'], "-", row['Преимущество'])
            else:
                er += 1
            if er == 2:
                print("Данные о препарате отсутствуют")
        if k==2:
            print(("Введите название препарата для получения расширенной информации о преимуществах:"))
            i=input().strip().lower()
            print()
            res = df[df['Продукт'].str.contains(i)]
            if len(res) > 0:
                for index, row in res.iterrows():
                    print(row['Продукт'], '(ПРОИЗВОДИТЕЛЬ:', row['Производитель'], 'ЛЕКФОРМА:',  row['Лекформа'], 'ПОКАЗАНИЕ:', row['Показание'] + ')')
                    print('ПРЕИМУЩЕСТВА:', row['Преимущество'])
            else:
                er += 1
            i=i[0].upper()+i[1:]
            print()
            res = df[df['Продукт'].str.contains(i)]
            if len(res) > 0:
                for index, row in res.iterrows():
                    print(row['Продукт'], '(ПРОИЗВОДИТЕЛЬ:', row['Производитель'], 'ЛЕКФОРМА:',  row['Лекформа'], 'ПОКАЗАНИЕ:', row['Показание'] + ')')
                    print('ПРЕИМУЩЕСТВА:', row['Преимущество'])
            else:
                er += 1
            if er == 2:
                print("Данные о препарате отсутствуют")
        if k==3:
            print("Спасибо, что воспользовались нашим справочником!")
            break
    except:
        print("Некорректный ввод. Попробуйте еще раз :)")
```
